```python
import numpy as np
from sklearn.datasets import load_iris

def getArtificialData(mean, cov, nx, nt):
    x = np.random.multivariate_normal(mean, cov, nx)
    t = np.random.multivariate_normal(mean, cov, nt)
    mu = np.mean(x, axis=0)
    Sigma = np.cov(x, rowvar=False)
    return x, t, mu, Sigma

def evaluateMultiVarGauss(t, mu, Sigma):
    d = t.shape[1]
    p = np.zeros(t.shape[0])
    detSigma = np.linalg.det(Sigma)
    invSigma = np.linalg.inv(Sigma)
    for i in range(t.shape[0]):
        diff = t[i] - mu
        p[i] = np.exp(-0.5 * np.dot(np.dot(diff, invSigma), diff)) / ((2 * np.pi) ** (d / 2) * np.sqrt(detSigma))
    return p

def calculateTestSet(t, pXY, yEx):
    misclassified = 0
    for i in range(len(t)):
        max_prob_index = np.argmax(pXY[i])
        print(f"Test point {i+1}: \n Actual label: {yEx[i]} \n Predicted label: {max_prob_index} \n  Probabilities: {pXY[i]}")
        if max_prob_index != yEx[i]:
            misclassified += 1
    print(f"Number of misclassified points: {misclassified}")

def calculateTiedCovariance(data):
    N = data.shape[0]
    mu = np.mean(data, axis=0)
    diff = data - mu
    Sigma_hat = (1 / N) * np.dot(diff.T, diff)
    return Sigma_hat

# Task 1
mean1 = [0, 0]
cov1 = [[2, 1], [1, 50]]
nx1 = 40
nt1 = 10

mean2 = [7, 5]
cov2 = [[3, 1], [1, 3]]
nx2 = 80
nt2 = 20

mean3 = [5, 5]
cov3 = [[5, 2], [2, 3]]
nx3 = 20
nt3 = 5

x1, t1, mu1, Sigma1 = getArtificialData(mean1, cov1, nx1, nt1)
x2, t2, mu2, Sigma2 = getArtificialData(mean2, cov2, nx2, nt2)
x3, t3, mu3, Sigma3 = getArtificialData(mean3, cov3, nx3, nt3)

merged_test_points = np.concatenate((t1, t2, t3), axis=0)

pXY1 = evaluateMultiVarGauss(merged_test_points, mu1, Sigma1)
pXY2 = evaluateMultiVarGauss(merged_test_points, mu2, Sigma2)
pXY3 = evaluateMultiVarGauss(merged_test_points, mu3, Sigma3)

pXY = np.column_stack((pXY1, pXY2, pXY3))

yEx = np.concatenate((np.zeros(nt1), np.ones(nt2), np.ones(nt3) * 2)).astype(int)

print("\nTask 1 Results:")
calculateTestSet(merged_test_points, pXY, yEx)
```

**SOLUTION-2**

```python
Sigma_hat1 = calculateTiedCovariance(np.concatenate((x1, x2, x3), axis=0))

pXY1_tied = evaluateMultiVarGauss(merged_test_points, mu1, Sigma_hat1)
pXY2_tied = evaluateMultiVarGauss(merged_test_points, mu2, Sigma_hat1)
pXY3_tied = evaluateMultiVarGauss(merged_test_points, mu3, Sigma_hat1)

pXY_tied = np.column_stack((pXY1_tied, pXY2_tied, pXY3_tied))

print("\nTask 2 Results (Tied Covariances):")
calculateTestSet(merged_test_points, pXY_tied, yEx)
```

```python
def getIrisData():
    iris = load_iris()
    data = iris.data
    target = iris.target
    x_train = np.concatenate((data[:40], data[50:90], data[100:140]), axis=0)
    x_test = np.concatenate((data[40:50], data[90:100], data[140:150]), axis=0)
    return x_train, x_test

x_train_iris, x_test_iris = getIrisData()

mu1_iris = np.mean(x_train_iris[:40], axis=0)
mu2_iris = np.mean(x_train_iris[40:80], axis=0)
mu3_iris = np.mean(x_train_iris[80:120], axis=0)
Sigma1_iris = np.cov(x_train_iris[:40], rowvar=False)
Sigma2_iris = np.cov(x_train_iris[40:80], rowvar=False)
Sigma3_iris = np.cov(x_train_iris[80:120], rowvar=False)

pXY1_iris = evaluateMultiVarGauss(x_test_iris, mu1_iris, Sigma1_iris)
pXY2_iris = evaluateMultiVarGauss(x_test_iris, mu2_iris, Sigma2_iris)
pXY3_iris = evaluateMultiVarGauss(x_test_iris, mu3_iris, Sigma3_iris)

pXY_iris = np.column_stack((pXY1_iris, pXY2_iris, pXY3_iris))

yEx_iris = np.concatenate((np.zeros(10), np.ones(10), np.ones(10) * 2)).astype(int)

print("\nTask 3 Results (Iris Dataset):")
calculateTestSet(x_test_iris, pXY_iris, yEx_iris)
```
